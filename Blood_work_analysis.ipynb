```python

from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage, SystemMessage
import os

load_dotenv(dotenv_path=r"C:\Users\leksh\Documents\AI_projects\.venv\.env")

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    groq_api_key=os.getenv("GROQ_API_KEY"),
    temperature=0
)

with open(r"C:\Users\leksh\Documents\AI_projects\Heathcare_App\blood_work.txt", "r") as f:
    blood_report = f.read()

prompt = f"""
You are a medical data extraction analyst.

Extract all blood test parameters from the report below.
For each parameter, return:
- Parameter name
- Patient value
- Normal range
- Status: High / Low / Normal

Then summarize which parameters need urgent attention.

Blood Report:
{blood_report}
"""

response = llm.invoke([
    SystemMessage(content="You are a medical data extraction analyst. Be precise and structured as a table."),
    HumanMessage(content=prompt)
])

print(response.content)
```
