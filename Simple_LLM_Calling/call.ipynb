```python
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage, SystemMessage
import os

load_dotenv(dotenv_path=r"C:\Users\leksh\Documents\AI_projects\.venv\.env")

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    groq_api_key=os.getenv("GROQ_API_KEY"),temperature =0
)

response = llm.invoke([
    SystemMessage(content="you are a friendly assistant and response should be less than 12 words"),
    HumanMessage(content="Give me a shopping list for an english dinner")
])

print(response.content)
```
