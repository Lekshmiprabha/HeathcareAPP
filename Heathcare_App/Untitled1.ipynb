```python
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage, SystemMessage
import os

load_dotenv(dotenv_path=r"C:\Users\leksh\Documents\AI_projects\.venv\.env")

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    groq_api_key=os.getenv("GROQ_API_KEY"),
    temperature=0
)

with open(r"C:\Users\leksh\Documents\AI_projects\Heathcare_App\blood_work.txt", "r") as f:
    blood_report = f.read()

diet_prompt = f"""
You are an expert Indian dietician and nutritionist.

Based on the blood report below, suggest a detailed Indian diet plan that:
- Addresses abnormal parameters
- Includes specific Indian foods to EAT and AVOID
- Covers breakfast, lunch, dinner and snacks
- Uses commonly available Indian ingredients
- Mentions portion sizes where relevant

Blood Report:
{blood_report}
"""

response = llm.invoke([
    SystemMessage(content="You are an expert Indian dietician. Give practical, specific Indian food recommendations only less than 100 words."),
    HumanMessage(content=diet_prompt)
])

print(response.content)
```
